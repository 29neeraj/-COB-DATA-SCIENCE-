```python
df['duration'] = df['duration'].replace('[^0-9]', '', regex=True)
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['release_year'], color='skyblue')
plt.title('Box Plot of Release Year')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df['is_movie'] = df['type'] == 'Movie'
df.loc[df['is_movie'], 'duration'] = df.loc[df['is_movie'], 'duration'].replace(' min', '', regex=True).astype(float)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='type', y='release_year', data=df)
plt.title('Box Plot of Release Year')
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['is_movie'] = df['type'] == 'Movie'

df.loc[df['is_movie'], 'duration'] = df.loc[df['is_movie'], 'duration'].replace(' min', '', regex=True).astype(float)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='type', y='release_year', data=df)
plt.title('Box Plot of Release Year')

plt.subplot(1, 2, 2)
sns.boxplot(x='type', y='duration', data=df)
plt.title('Box Plot of Duration (minutes)')

plt.tight_layout()
plt.show()
```
